```python
import pandas as pd
import json
from datetime import datetime
```

```python
data_path = "/Users/alex/Library/CloudStorage/OneDrive-MUNI/Dokumenty/Projects Data/GNNs/cz_traffic/"
#data_path = "/media/sf_bely/OneDrive - MUNI/Dokumenty/Projects Data/GNNs/cz_traffic/"
```

```python
%%time

traffic_file_name = data_path + 'traffic-2022-02.ndjson'
#data_df = pd.read_json(traffic_file_name, lines=True) #out of memory
fields = ['timestamp', 'segment_id', 'segment_name', 'segment_road_name', 'lat', 'lon', 'intensity', 'avg_speed', 'speed_variance']
df_dict = {field: [] for field in fields}
with open(traffic_file_name) as f:
    for line in f:
        record = json.loads(line)
        df_dict['timestamp'].append(datetime.fromisoformat(record['timestamp'][:-1]))
        df_dict['segment_id'].append(int(record['segment_id']))
        df_dict['segment_name'].append(record['segment_name'])
        df_dict['segment_road_name'].append(record['segment_road_name'])
        df_dict['lat'].append(float(record['segment_location']['lat']))
        df_dict['lon'].append(float(record['segment_location']['lon']))
        df_dict['intensity'].append(int(record['intensity']))
        df_dict['avg_speed'].append(float(record['avg_speed']))
        df_dict['speed_variance'].append(float(record['speed_variance']))
```

```python
len(df_dict['segment_id'])
```

```python
data_df = pd.DataFrame.from_dict(df_dict)
data_df.head()
```

```python
data_df.to_parquet(data_path + 'traffic-2022-02.pqt')
```
