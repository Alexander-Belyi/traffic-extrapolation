```python
import os
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')


%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
df = pd.DataFrame()

path = '../tlc/results/'

for m in os.listdir(path):
    with open(path + m) as f:
        res = json.load(f)
    df[m.split('.')[0]] = pd.Series(res)

plt.figure(figsize=(20,6))
plt.ylim((0.5, 1))
plt.ylabel('r2 score')
sns.lineplot(data=df).set_title('TLC')
plt.legend()
plt.show()
```

```python
df = pd.DataFrame()

path = '../mta/results/'

for m in os.listdir(path):
    with open(path + m) as f:
        res = json.load(f)
    df[m.split('.')[0]] = pd.Series(res)

plt.figure(figsize=(20,6))
plt.xticks(rotation=45)
plt.ylim((0.0, 1))
plt.ylabel('r2 score')
sns.lineplot(data=df).set_title('MTA')
plt.legend()
plt.show()
```
